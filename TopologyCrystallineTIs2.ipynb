# Definitions of Functions for Berry Curvature and BHZ Model

## Berry Curvature for Hamiltonian without spin-mixing

I realized that the calculation of Berry quantities of one member Kramer pair becomes feasible if there is no mixing term in the spin space. Then, the bloch wave functions of tr pairs will be 2-component spinors wavefunction with either up or down spin polarization. This way, only looking the up or down spinor wavefunction, we can form one member of Kramer pair in the entire BZ. This is what is done in Jaanen paper, [see](https://arxiv.org/abs/1108.3337). Even though I read this model in Bernevig's paper first, realized this particular property of the model in this paper after seeing the citation paper by Jasper et al. Let's first attack the problem with the hamiltonian that has no spin mixing term, then complicate it by including spin mixing. In the first case, we will first check the consistency coondition due to line invariants and look at the locations of vortices by calculation evolution of Berry phase and Berry curvature distribution on the BZ.

The first model of consideration is qw-spin hall insulator by berneving et al. The phases are as follows:

* For $\frac{M}{B} < 0 $: no topological defect
* For $0 < \frac{M}{B} < 4 $: topological defect at $\Gamma$
* For $4 < \frac{M}{B} < 8 $: topological defect at X and Y
* For $8 < \frac{M}{B} $: no topological defect 

To find the distribution of Berry quantities, we start by implementing the code for Berry curvature. It is important to note that this code will not be general in the following sense. We will look at the Hamiltonians that have no spin mixing term.Since we have **U(2)** freedom at every point in BZ due to degeneracy and random phases, we need to impose the condition that Bloch Wavefunction's are spin polarized by hand. This is done by choosing Bloch wf's with lower or upper 2 components to be zero, which is possible due block form of the Hamiltonian. The discrete Berry curvature will be calculated by following the paper, [see](https://arxiv.org/abs/cond-mat/0503172).

```python
def Berry_curv(Ham, var_vec, N):

    '''
    Params:
    -------
        Ham: the hamiltonian function which will return the Bloch Hamiltonian
        var_vec: the set of constants in the Ham
        N: the number of points along one axis. For example, for 2d there are N^2 points in the entire BZ.
    '''

    #discrete x and y-axis
    kx_int, ky_int = np.linspace(-np.pi, np.pi, N), np.linspace(-np.pi, np.pi, N)
    
    # k_mesh which stores the Bloch wavefunctions on these points
    k_mesh = np.empty(shape=(N,N), dtype="object")          

    # step = dk_x = dk_y
    step = 2*np.pi/(N-1)                                 
    # N-1 x N-1 array which stores Berry curvature in the dual lattice sites.
    berry_curv = np.empty(shape=(N-1,N-1),dtype=np.float32)              

    for i, k1 in enumerate(kx_int):
        for j, k2 in enumerate(ky_int):
            H = Ham((k1, k2), var_vec)[:2, :2]               
            evals, ekets = la.eigh(H)
            
            #spinor bloch wf with nonzero upper components
            half1 = np.append(ekets, np.zeros((2,2)), axis=0)             
            H = Ham((k1, k2), var_vec)[2:, 2:]
            evals, ekets = la.eigh(H)
            
            #spinor bloch wf with nonzero lower components
            half2 = np.append(np.zeros((2,2)), ekets, axis=0)             
            k_mesh[i,j] = np.append(half1, half2, axis=1)



    # apply pbc
    k_mesh[:,-1] = k_mesh[:,0]
    k_mesh[-1,:] = k_mesh[0,:]


    for i, k1 in enumerate(kx_int[:-1]):
        for j, k2 in enumerate(ky_int[:-1]):
            
            # Calculate berry phase over individual edges of plaquette
            U1 = np.vdot(k_mesh[i,j][:,0], k_mesh[i, j+1][:,0])
            U2 = np.vdot(k_mesh[i,j+1][:,0], k_mesh[i+1, j+1][:,0])         
            U3 = np.vdot(k_mesh[i+1, j+1][:,0], k_mesh[i+1, j][:,0])
            U4 = np.vdot(k_mesh[i+1, j][:,0], k_mesh[i, j][:,0])
            
            # value of flux over the small plaquette
            B = U1* U2* U3 * U4                                             
            berry_curv[i,j] = np.real(np.angle(B))/(2*np.pi)/step**2

    return berry_curv
```

We can make a few commments. Far in the trivial state, we have the configuration where at the $\Gamma$ and $M$ points, there is a negative source of Berry curvature. At $X$ and $Y$ points, there is a positive source of Berry curvature. As the ration $\frac{M}{B}$ goes to $0$, the positive source gets smeared and becomes constant positive berry curvature for some region, indicated by darker green. The negative source of berry curvature at $M$ gets smoother as well. However, the source at $\Gamma$ gets even more localized, as it becomes a point in the image. This is reasonable since at $\frac{M}{B}=0$, there is a gap closing and phase transition at $\Gamma$.

In addition, this configuration of pink being antivortex and green being vortex is consistent with the values of line invariants given above.

I start with a function which calculates the eigenvalues of plaquette operator on the BZ.


```python
def Berry_curv2(Ham, ham_pars, N):

  '''
  Ham: the hamiltonian function which will return the Bloch Hamiltonian
  var_vec: the set of constants in the Ham
  N: the number of points along one axis. For example, for 2d there are N^2 points in the entire BZ.
  '''
  # discrete x-axis
  kx_int = np.linspace(-np.pi, np.pi, N)    
  # discrete y-axis
  ky_int = np.linspace(-np.pi, np.pi, N)  
  # step = dk_x = dk_y
  step = 2*np.pi/(N-1)      
  # N-1 x N-1 array which stores Berry curvature in the dual lattice sites.
  berry_curv1 = np.zeros(shape=(N-1,N-1))              
  berry_curv2 = np.zeros(shape=(N-1,N-1)) 

  M = 3
  for i, k1 in enumerate(kx_int[:-1]):
    for j, k2 in enumerate(ky_int[:-1]):
      k_plaq=[(k1, k2), (k1+step, k2), (k1+step, k2+step), (k1, k2+step)]
      B = band.Z2_invariant(k_plaq, Ham, ham_pars, [0,1]) 
      B1 = max(B)                                         #value of flux over the small plaquette
      B2 = min(B)
      berry_curv1[i,j] = B1/step**2
      berry_curv2[i,j] = B2/step**2


  return berry_curv1, berry_curv2
```

To obtain the continuous configuation of eigenvalues, I will use the following recipe.

1.   First extract the maximum eigenvalue on each row
2.   Starting on the row with maximum max, proceed from this region of maximum go forward and backwards to obtain a continuous function on each row
3.   Determine the function to which the row belongs by comparing the point-wise distance between each row. Closes ones will be continuation of each other.
4.   Do this for only lower half and determined the other half by TR symmetry.

Now, we observe that the for example for $\frac{M}{B} =2$, even thought the system in topologically non-trivial phase, the vortex at the origin, which was localized for $\frac{M}{B} =0^+$, is smeared around the origin. This is reasonable since these kramer pairs are degenerate in the entire BZ. Hence, they don't form Dirac cone around the TR invariant point. Hence, the argument by Jan wezel paper does not apply. To see the relevant case for this paper let's add a term without breaking inversion symmetry/ opening-closing gap. This term will have spin-mixing to form Dirac cone regions.

One can check that the Hamiltonian with such properties is given by

```python
def HamBHZ2(k_vec, ham_pars):

    kx = k_vec[0]
    ky = k_vec[1]
    A = ham_pars['A']
    B = ham_pars['B']
    C = ham_pars['C']
    D = ham_pars['D']
    M = ham_pars['M']
    delta = ham_pars['delta']
    b1 = ham_pars['b1']
    b2 = ham_pars['b2']

    
    E_k = C- 2*D*(2- np.cos(kx) - np.cos(ky))
    d_1 = A*np.sin(kx)
    d_2 = A*np.sin(ky) 
    d_3 = -2*B*(2 -M/(2*B)-np.cos(kx) - np.cos(ky))
    H = E_k*band.S00 + d_1*band.Szx + d_2*band.S0y + d_3*band.S0z
    H_int = delta*(-band.Syy)                         #term breaking the inversion symmetry
    H_int1 = b1*(np.sin(kx))*band.Sxx + b2*(np.sin(ky))*band.Syx

    return H + H_int + H_int1

```

where H_int1 is the new spin-mixing term. We will start by increasing $b_1$ and $b_2$ by a very small amount while checking it does not modify the gap. To check this, we will use gap function

```python
def gap(Ham, var_vec, N):
  kx_path = np.linspace(-np.pi, np.pi, N)
  ky_path = np.linspace(-np.pi, np.pi, N)
  H = Ham((0,0), var_vec)
  n = H.shape[0]
  evals = la.eigh(H)[0]
  gap = np.abs(evals[n//2]- evals[n//2 -1])
  for kx in kx_path:
    for ky in ky_path:
      evals = la.eigh(Ham((kx,ky), var_vec))[0]
      gap_n = np.abs(evals[n//2]- evals[n//2 -1])
      if gap_n < gap:
        gap = gap_n
  return gap
```

```python
A=1    #rescaling the energy
C= 0               #energy term proportional to identity
D= 0               #energy term proportional to identity
delta=0        #no mixing in the spin space
B=1
b1 = 0.04
b2 = -0.04

N=100
num = 80
gap_list = []
M_list = np.linspace(-1,9, num)
for i in M_list:
  M = i
  var_list = {'A': A, 'B': B, 'C': C, 'D': D, 'M': M, 'delta': delta, 'b1':b1, 'b2': b2}
  gap_list.append(gap(HamBHZ2, var_list, N))



fig = plt.figure( figsize =(14,4))
ax1 = fig.add_subplot(111)
ax1.scatter(M_list, gap_list, c='blue', s=6, label='gap')

plt.legend(loc='upper left');
plt.grid(True)
plt.show()
```
